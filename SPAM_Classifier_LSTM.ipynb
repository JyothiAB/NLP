```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/JyothiAB/NLP/main/train1.csv')
```

```python
df2 = pd.read_csv('https://raw.githubusercontent.com/JyothiAB/NLP/main/train2.csv')
df3 = pd.read_csv('https://raw.githubusercontent.com/JyothiAB/NLP/main/train3.csv')
df4 = pd.read_csv('https://raw.githubusercontent.com/JyothiAB/NLP/main/train4.csv')
df5 = pd.read_csv('https://raw.githubusercontent.com/JyothiAB/NLP/main/train5.csv')
```

```python
df = pd.concat([df1,df2,df3,df4,df5], axis=0)
```

```python
df
```

```python
X = df[['title','text']]
y = df['label']
```

```python
X.shape
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import Embedding
```

```python
#Data preprocessing
message = X.copy()
```

```python
message.reset_index(inplace=True)
```

```python
message
```

```python
import nltk
import re

from nltk.corpus import stopwords
from nltk import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
lm = WordNetLemmatizer()
title = []
for i in range(0,len(message)):
  review_title = re.sub('[^a-zA-Z]', ' ', str(message['title'][i]))
  review_title = review_title.lower()
  review_title = review_title.split()

  review_title = [lm.lemmatize(words) for words in review_title if not words in stopwords.words('english')]
  review_title = ' '.join(review_title)

  title.append(review_title)
print(title)                  
```

```python
#One hot representation
title_one_hot = [one_hot(words,5000) for words in title]
print(title_one_hot)
```

```python
#Padding
title_pad = pad_sequences(title_one_hot,maxlen=20, padding='pre')
print(title_pad)
```

```python
#Training and Testing Data from actual data
from sklearn.model_selection import train_test_split
x_training,x_test,y_training,y_test = train_test_split(title_pad,y,test_size=0.1,random_state=20)
```

```python
#Training and Validation Data from Training data
x_train, x_val, y_train, y_val = train_test_split(x_training,y_training,test_size=0.1,random_state=30)
```

```python
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras import optimizers

def model_builder(hp):
  model = Sequential()
  model.add(Embedding(input_dim=5000, output_dim=40, input_length=20))
  model.add(LSTM(units=hp.Int('units_', min_value=32, max_value=512, step=32), activation='relu'))
  model.add(Dense(1,activation='sigmoid'))

  # Tune the learning rate for the optimizer
  # Choose an optimal value from 0.01, 0.001, or 0.0001
  hp_learning_rate = hp.Choice('learning_rate', values=[0.01, 0.001, 0.0001])
  model.compile(optimizer= optimizers.Adam(learning_rate=hp_learning_rate),
                loss='binary_crossentropy',
                metrics=['accuracy'])

  return model
```
